```python
df = pd.read_csv("10000 Sales Records.csv")
df
```

```python


# 1.1) Построить любой график (любого рода) с помощью чистого matplotlib / pylab
# 1.2) Построить несколько суб-графиков на одном графике используя .subplot.
# Графики должны быть разных типов и хотя бы два из них должны быть связаны с вашим датасетом. 
# Хотя бы в одном графике подпишите оси. Во всех графиках подписать название графика.
plt.subplot(2, 2, 1)
plt.hist(df['Region'], bins=50)
plt.title('Кол-во заказов в регионах')

plt.subplot(2, 2, 3)
plt.scatter(df['Unit Price'], df['Item Type'])
plt.xlabel('Unit price')
plt.ylabel('Item type')

plt.subplot(2, 2, 4)
df1=df.head(500)
plt.bar(df1['Sales Channel'].unique(), df1['Sales Channel'].value_counts())
plt.xlabel('Target')
plt.ylabel('Count')
plt.title('Кол-во рнлайн и оффлайн заказов')
plt.show()
```

```python
# 2) На основе Вашего датасета построить три разных вида графиков используя 
# pandas.DataFrame.plot / Series.plot (мануал) 
# Среди графиков должен присутствовать один boxplot 
# (обязательно надо использовать параметр by, для исследования взаимосвязи числовой переменной и какой-либо другой). 
# Графики должны быть построены как через методы DataFrame, так и через методы Series. 
# То есть оба варианта нужно продемонстрировать.


df3=df.head(100)
# Boxplot с использованием DataFrame.plot
df1.plot(kind='box', by='Order Priority', figsize=(8, 6))
plt.xlabel('Total_crimes')
plt.ylabel('Serious')
plt.title('Boxplot by Order priority')
plt.show()
# График распределения числовой переменной с использованием Series.plot
df3['Units Sold'].plot(kind='hist', bins=10, figsize=(8, 6))
plt.xlabel('units')
plt.ylabel('Count')
plt.title(' Count of units')
plt.show()
# График взаимосвязи двух числовых переменных с использованием DataFrame.plot
df3.plot(kind='scatter', x='Units Sold', y='Total Profit', figsize=(8, 6))
plt.xlabel('Units Sold')
plt.ylabel('Total Profit')
plt.title('Total profit by units sold')
plt.show()
```

```python
# 3) Seaborn
# 3.1) Построить .pairplot с помощью seaborn на каком-либо наборе данных, 
# проанализировав взаимосвязь набора числовых переменных


sns.pairplot(df)
plt.show()
```

```python
df2=df.head(100)
numeric_vars = ['Total Revenue', 'Total Cost', 'Total Profit']
sns.pairplot(df1[numeric_vars])
plt.show()
```

```python
# 3.2) Построить .jointplot
x_var = 'Total Revenue'
y_var = 'Order Priority'
sns.jointplot(data=df2, x=x_var, y=y_var, kind='scatter')
plt.show()
```

```python
# 3.3) Построить .boxplot или .violinplot на каком-либо наборе данных, 
# проанализировав какой-то числовой показатель в разрезе категориального
numeric_var = 'Units Sold'
categorical_var = 'Item Type'
sns.boxplot(data=df2, x=categorical_var, y=numeric_var)
plt.xlabel(categorical_var)
plt.ylabel(numeric_var)
plt.title('Boxplot of {} by {}'.format(numeric_var, categorical_var))
plt.show()
```

```python
#3.4) Построить график heatmap попарной корреляции всех числовых признаков 
# (категориальные признаки убираете из графика)
numeric_features = df2.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_features.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
